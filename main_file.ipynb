```python
import random
import pandas as pd

selected_teachers = []
selected_teachers = InstructorsDF['Names'].tolist()

selected_courses = []
selected_courses = CoursesDF['Course Code'].tolist()
#print("Selected Courses: ", selected_courses)
#print("Selected Courses: ", len(selected_courses))
courses = selected_courses.copy()
#print("Courses: ", courses)
#print("Courses: ", len(courses))
random.shuffle(courses)
Allocated_Teacher_and_courses = []

for teacher in selected_teachers:
    teacher_courses = []
    
    for _ in range(2):
        if courses:
            course = courses.pop(0)
            teacher_courses.append(course)
    if teacher_courses:
        Allocated_Teacher_and_courses.append({'Teacher': teacher, 'Course Code': teacher_courses})

for i in Allocated_Teacher_and_courses:
    print(i)
```

```python
# Assumption: The block "C" has 6 floors and exams are held on 2nd, 3rd and 4th floor
#Assumption: each floor has 10 classrooms
# Assumption: classes/exams are to be held from 9:00am to 5:00pm
classrooms = [] # empty list for rooms
timings = [] # empty list for timings
# Define the floors on which classes are held
class_floors = [2, 3, 4] # suppose classes
minute_ending = 00 # there will be a 10 minute break after each exam/class
for floor in range(1,7):
    if floor in class_floors:
        for room_number in range(1, 11): # room numbers start from 1 to 10
            room = f"c-{floor}{room_number:02d}"
            classrooms.append(room)
hour = 9 # First exam or class will be held at 9:00 and till 16:10
while (hour < 17):

    time= f"{hour:02d}:{minute_ending:02d}-{(hour + 1):02d}:{minute_ending:02d}"
    if minute_ending == 50:
        hour += 1
    minute_ending += 10
    minute_ending =  minute_ending % 60
    hour += 1
    if minute_ending == 60:
        minute_ending = 0
        print("hour is: ", hour)
    timings.append(time)
print("Available rooms: ", classrooms)
print("Number of Available rooms: ", len(classrooms) )
print("Timings: ", timings)
print("Number of Time Slots: ", len(timings))

Days = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
print('Days', Days)


```

```python
import pandas as pd
from datetime import datetime, timedelta
import random

TeacherName = ''
CourseCode = ''
TimeSlot = ''
RoomNumber = ''
Day = ''
DayN = ''
Date = datetime.now().date() # today's date
date_tobeusedinloop = Date


#package = TeacherName,CourseCode,TimeSlot,RoomNumber,Day,date_tobeusedinloop
selected_packages = []
# in this list we will append the package 
# this list will help us to keep track that we don't have clashes
track_of_timeslots_and_date = {
    'Date' : None,
    'Time Slots' : [None,None,None,None,None,None,None]
    
}
# this dictionary will help me to know that for one Date all time slots have been utilised
idx = 0
temp_time_list = []
for i in range(len(Allocated_Teacher_and_courses)):
    # Extract the teacher 
    TeacherNCourses = Allocated_Teacher_and_courses[i]
    TeacherName = TeacherNCourses['Teacher']
    CourseCodes = TeacherNCourses['Course Code']
    for coursecode in CourseCodes:
        #print("IDX: ", idx)
        # for each course code allocated to one Teacher we will assign it a date, time,room,day
        if idx >=8: # idx >= 8 as we increment by 2 so after using the 6th index we increment by two and get out of range index
            idx = 0
            date_tobeusedinloop +=  timedelta(days=1)
            # increment the date
        NewCourseCode = coursecode
        NewTimeSlot = timings[idx]
        idx += 2 
        track_of_timeslots_and_date['Date'] = date_tobeusedinloop
        temp_time_list.append(NewTimeSlot)
        track_of_timeslots_and_date['Time Slots'] = temp_time_list
        # incrementing by 2 so that consecutive slots aik teacher na invigilate karay
        # consecutive slot 
        NewRoomNumber = random.choice(classrooms)
        # date tub change hu gi agr aik din k saray slots booked hein
        NewDayN = date_tobeusedinloop.weekday()
        NewDay = Days[NewDayN]
        # now putting a check for Friday, Saturday, Sunday
        # Hard Constraint that exam cannot be held on Weekends and Soft Constraint that on Friday break between 1-2
        #(based on out schedule that would be no exam during slots in excluded_slots )
        if NewDay in ['Friday','Saturday','Sunday']:
            
            if NewDay == 'Friday':
                excluded_slots = ['12:30-13:30', '13:40-14:40']
                
                while NewTimeSlot in excluded_slots:
                    #print("IDX here: ", idx)
                    idx -= 1
                    #print("IDX Now: ", idx)
                    NewTimeSlot = timings[idx]
                    idx += 1
                        
            elif NewDay == 'Saturday':
                date_tobeusedinloop = date_tobeusedinloop + timedelta(days=2)
                track_of_timeslots_and_date['Date'] = date_tobeusedinloop
                day_of_week = date_tobeusedinloop.weekday()
                NewDay = Days[day_of_week]
            
            else:
                # if Day is 'Sunday' then we just have to increment date by one day
                date_tobeusedinloop = date_tobeusedinloop + timedelta(days=1)
                track_of_timeslots_and_date['Date'] = date_tobeusedinloop
                day_of_week = date_tobeusedinloop.weekday()
                NewDay = Days[day_of_week]
            
        package = TeacherName, NewCourseCode, NewTimeSlot, NewRoomNumber, NewDay, date_tobeusedinloop
        # yaha pe you will check kay previous packages mein same content to nhi hay
        
        for p in selected_packages:
            prevTeacher, PrevCourseCode, PrevTimeSlot, PrevRoom, PrevDay, PrevDate = p
            # check 1 consecutive time slots pe aik teacher invigilation nhi kar sakta
            if date_tobeusedinloop==PrevDate:
                    if NewTimeSlot==PrevTimeSlot:
                        if NewRoomNumber==PrevRoom:
                            # mean same timeslot same room tu we need to change it
                            while(NewRoomNumber!=PrevRoom):
                                # to avoid repeat what if random kar k dubara woi a jaye
                                NewRoomNumber = random.choice(classrooms)
                                
        selected_packages.append(package)
        
        
# for p in selected_packages:
#     TeacherName,CourseCode,TimeSlot,RoomNumber,Day,date_tobeusedinloop = p
#     print('Teacher Name-->',TeacherName)
#     print('CoursesCode-->',CourseCode)
#     print('TimeSlot-->',TimeSlot)
#     print('Room Number-->', RoomNumber)
#     print('Day-->',Day)
#     print('Date-->',date_tobeusedinloop)
#     print('\n')
    
# print('Total Exams: ',len(selected_packages))
print(selected_packages)
```

## APPLICATION OF GENETIC ALGORITHM

First of all we are going to generate 3 more timetables (totally randomised). After that we will make population and choromosomes for these timetables.


Population = TimeTable 1.

```python
# Generating 3 timetables of student and teacher
import random
import pandas as pd
from datetime import datetime, timedelta
Days = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
# timings
# classrooms
# using all those previous lists we will generate three random timetable
for counter_itr in range(0,4):
    selected_teachers = []
    selected_teachers = InstructorsDF['Names'].tolist()
    selected_courses = []
    selected_courses = CoursesDF['Course Code'].tolist()
    courses = selected_courses.copy()
    random.shuffle(courses)
    Allocated_Teacher_and_courses = []
    TeacherName2 = ''
    CourseCode2 = ''
    TimeSlot2 = ''
    RoomNumber2 = ''
    Day2 = ''
    DayN2 = ''
    Date2 = datetime.now().date() # today's date
    date_tobeusedinloop2 = Date

    genetic_dataframe = {
        
        'Date' : [],
        'Day' : [],
        'Time Slot' : [],
        'Room Number' : [],
        'Teacher Name' : [],
        'Course Code' : [],
        'Student Name' : []   
    }

    track_of_timeslots_and_date = {
        'Date' : None,
        'Time Slots' : [None,None,None,None,None,None,None]
        
    }

    # aik student jetney courses le raha hay
    # subse pehle generate random timetable for teacher
    # baad mein we will merge it with student and onkay courses
    # for _ in range(len(CoursesDF['Course Code'].unique())):
    for teacher in selected_teachers:
        teacher_courses = []
        
        length1 = random.choice(range(0,2))
        for _ in range(length1):
            if courses:
                course = courses.pop(0)
                teacher_courses.append(course)

        Allocated_Teacher_and_courses.append({'Teacher': teacher, 'Course Code': teacher_courses})
    # this way kysi teacher ku aik course assign huga kysi ku du and kysi ku koi bhe nahi

    selected_packages2 = []
    # now the next step is randomly assign room number date,day and time to each teacher to make teacher timetable
    idx = 0
    temp_time_list = []
    for i in range(len(Allocated_Teacher_and_courses)):
        tNc =  Allocated_Teacher_and_courses[i]
        teachername = tNc['Teacher']
        coursesname = tNc['Course Code']
        
        for course in coursesname:
            
            # because there are some teachers who are not assigned with any course in that case just break the loop
            if course=='':
                break
            
            # first of all randomly fill the slots of one date 
            if idx >= 8:
                idx = 0
                date_tobeusedinloop2 += timedelta(days=1)
            
            Newcc = course
            newtime = timings[idx]
            # Handling the constrait that one Teacher should not invigilate two exams in a row
            idx += 2
            track_of_timeslots_and_date['Date'] = date_tobeusedinloop2
            temp_time_list.append(NewTimeSlot)
            track_of_timeslots_and_date['Time Slots'] = temp_time_list
            # to keep track k aik din k saray slots fill hun
            NewRoomNumber = random.choice(classrooms)
            # date tub change hu gi agr aik din k saray slots booked hein
            NewDayN = date_tobeusedinloop2.weekday()
            NewDay = Days[NewDayN]
            
            package = teachername , Newcc, newtime, NewRoomNumber, NewDay, date_tobeusedinloop2
            selected_packages2.append(package)
            
    # 100% random timetable for teacher has been generated
    Invigilation_assignments2 = { "Teacher":[],
                    "Course Code": [],
                    "Time Slot": [],
                    "Room Number": [],
                    "Day": [],
                    "Date": []}       
    for p in selected_packages2:
        TeacherName2,CourseCode2,TimeSlot2,RoomNumber2,Day2,date_tobeusedinloop2 = p
        Invigilation_assignments2["Teacher"].append(TeacherName2)
        Invigilation_assignments2["Course Code"].append(CourseCode2)
        Invigilation_assignments2["Time Slot"].append(TimeSlot2)
        Invigilation_assignments2["Room Number"].append(RoomNumber2)
        Invigilation_assignments2["Day"].append(Day2)
        Invigilation_assignments2["Date"].append(date_tobeusedinloop2)
        
        

    Invigilation_assignmentsDF2 = pd.DataFrame(Invigilation_assignments2)
    grouped = Invigilation_assignmentsDF2.groupby("Course Code")

    # Now we will generate timetable of teacher with students  
    Teachers_list = []
    rooms = []
    Dates = []
    daysss =[]
    slotsss = []
    names = []
    ccc = []
    name_idx = 0
    for courseid in Eligible_Students["Course Code"]:
        if courseid in Invigilation_assignments2["Course Code"]:
            idx = Invigilation_assignments2["Course Code"].index(courseid)
            ccc.append(courseid)
            Teachers_list.append(Invigilation_assignments2["Teacher"][idx])
            rooms.append(Invigilation_assignments2["Room Number"][idx])
            Dates.append(Invigilation_assignments2["Date"][idx])
            daysss.append(Invigilation_assignments2["Day"][idx])
            slotsss.append(Invigilation_assignments2["Time Slot"][idx])
            names.append(Eligible_Students['Student Name'].iloc[name_idx])
            name_idx += 1
            
    genetic_dataframe["Teacher Name"] = Teachers_list
    genetic_dataframe['Room Number'] = rooms
    genetic_dataframe['Date'] = Dates
    genetic_dataframe['Day'] = daysss
    genetic_dataframe["Time Slot"] = slotsss
    genetic_dataframe['Student Name'] = names
    genetic_dataframe['Course Code'] = ccc
    df = pd.DataFrame(genetic_dataframe)
    file_name = f"Timetable_{counter_itr+1}.csv"
    df.drop_duplicates(inplace = True)
    df.to_csv(file_name) 
        
        
# Now we have 4 timetables on which we will apply GA
# Each Timetable is a Gene
# [1011] --> is chromosome
# multiple choromosome make one population
df   
    

```

```python
def no_exam_on_weekends(temp_df):
    temp = pd.DataFrame({})
    # Retreive data jahan jahan Day == 'Saturday' / 'Sunday'
    temp = temp_df[(temp_df['Day'] == 'Saturday') | (temp_df['Day'] == 'Sunday')]
    #print(temp)
    if temp.empty:
        return True
    return False
#print(no_exam_on_weekends(dfr))            
     
```

```python
def Friday_break(temp_df):
    temp = pd.DataFrame({})
    # Break on Friday during 12:30-13:30 13:40-14:40 
    #(two hours as our scheduled is made in a way that only then students will be free from 1 to 2)
    temp = temp_df[(temp_df['Day'] == 'Friday') & ((temp_df['Time Slot'] == '12:30-13:30') | (temp_df['Time Slot'] =='13:40-14:40'))]
    if temp.empty:
        return True
    return False
#print(Friday_break(dfr))
```

```python
def no_student_slots_clash(temp_df):
    grouped_df = temp_df.groupby('Time Slot')
    for time, group_df in  grouped_df:
        # Check the number of unique student names
        unique_count = group_df['Student Name'].nunique()
        if unique_count == len(group_df['Student Name']):
            return True
    return False
# no_student_slots_clash(dfr)
```

```python
def start_end_time_check(temp_df):
    valid_slots = ['09:00-10:00', '10:10-11:10', '11:20-12:20', '12:30-13:30', '13:40-14:40', '14:50-15:50', '16:00-17:00']
    time_slots = temp_df['Time Slot'].unique()
     # Check if all time slots are in the list of valid time slots
    for slot in time_slots:
        if slot not in valid_slots:
            return False
    return True
```
